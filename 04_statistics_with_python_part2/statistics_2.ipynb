## 2.1 Binomial Distribution
***

A **binomial** event is characterized by the possibility of occurrence of only two categories. These added categories represent the entire sample space and are also mutually exclusive, i.e. the occurrence of one implies the non-occurrence of the other.

In statistical analysis, the most common use of the binomial distribution is to solve problems involving **success** and **failure** situations.

# $$P(k)=\binom{n}{k} p^k q^{n-k}$$

Where

$p$ = probability of success

$q = (1 - p)$ = probability of failure

$n$ = number of events studied

$k$ = number of desired events that will be successful

### Binomial Experiment

1. Realization of $n$ identical tests.

2. The tests are independent.

3. Only two results are possible, example: True or false; Heads or tails; Success or failure.

4. The probability of success is represented by $p$ and that of failure by $1-p=q$. These probabilities do not change from test to test.

### Binomial distribution average

The expected value or mean of the binomial distribution is equal to the number of experiments performed multiplied by the chance of the event occurring.

# $$m = n \times p$$

### Standard deviation of binomial distribution

The standard deviation is the output between the number of experiments, the probability of success and the probability of failure.

# $$\sigma = \sqrt{n \times p \times q}$$

Combinations: you need to select k items out of n

Number of combinations of $n$ objects, taken $k$ each time, is:

# $$C_{k}^{n} = \binom{n}{k} = \frac{n!}{k!(n - k)!}$$

Where

## $$n! = n\times(n-1)\times(n-2)\times...\times(2)\times(1)$$
## $$k! = k\times(k-1)\times(k-2)\times...\times(2)\times(1)$$

By definition

## $$0! = 1$$

## 2.2 Poisson Distribution
***

It is employed to describe the number of occurrences in a specific time or space interval. Events are characterized by the possibility of counting successes, but not the possibility of counting failures.

As examples of processes where we can apply Poisson's distribution we have the determination of the number of customers entering a store at a given time, the number of cars arriving at a drive-thru of a snack bar at lunchtime, the determination of the number of accidents recorded on a stretch of road etc.

With poisson distribution we can count the number of successes in a range, but it's impossible to calculate the number of failures. For example, I can count how many people visit a store in the time range of one hour, but I cannot count how many people don't visit the store in the same time range.

# $$P(k) = \frac{e^{-\mu}(\mu)^k}{k!}$$

Where:

$e$ = constant whose approximate value is é 2,718281828459045

$\mu$ = represents the average number of occurrences in a given time or space

$k$ = number of hits in the desired range

### Poisson Experiment

1. The probability of an occurrence is the same throughout the observed interval.

2. The number of occurrences in a given interval is independent of the number of occurrences in other intervals.

3. The probability of an occurrence is the same at intervals of equal length.

### Poisson distribution mean

# $$\mu$$

### Poisson Standard Deviation

# $$\sigma = \sqrt{\mu}$$

## 2.3 Normal Distribution
***

The normal distribution is one of the most used in statistics. It is a continuous distribution, where the frequency distribution of a quantitative variable has the shape of a bell and is symmetrical in relation to its mean.

![Normal](https://caelum-online-public.s3.amazonaws.com/1178-estatistica-parte2/01/img001.png)

### Important features

1. It's symmetrical around the mean;

2. The area under the curve corresponds to the proportion 1 or 100%;

3. The measures of central tendency (mean, median and mode) present the same value;

4. The extremes of the curve tend to infinity in both directions and, theoretically, never touch the $x$ axis;

5. The standard deviation defines the flatness and width of the distribution. Wider and flatter curves have higher standard deviation values;

6. The distribution is defined by its mean and standard deviation;

7. The probability will always be equal to the area under the curve, delimited by the lower and upper limits.

# $$f(x) = \frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Where:

$x$ = normal variable

$\sigma$ = standard deviation

$\mu$ = mean

The probability is obtained from the area under the curve, delimited by the specified upper and lower limits. An example can be seen in the figure below.

![alt text](https://caelum-online-public.s3.amazonaws.com/1178-estatistica-parte2/01/img002.png)


To obtain the above area, simply calculate the function integral for the given intervals. According to the equation below:

# $$P(L_i<x<L_s) = \int_{L_i}^{L_s}\frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$


https://www.youtube.com/watch?v=IoMdxDIJgVc


Where:

$x$ = normal variable

$\sigma$ = standard deviation

$\mu$ = mean

$L_i$ = inferior limit

$L_s$ = superior limit

### Standardized Tables

The standardized tables were created to make it easier to obtain the values of the areas under the normal curve and eliminate the need to solve defined integrals.

To consult the values in a standardized table we just have to transform our variable into a standardized variable $Z$.

This variable $Z$ represents the departure in standard deviations of a value of the original variable from the mean.

# $$Z = \frac{x-\mu}{\sigma}$$

Where:

$x$ = normal variable with mean $\mu$ and standard deviation $\sigma$

$\sigma$ = standard deviation

$\mu$ = mean

### Building a standardized table
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

standardized_normal_table = pd.DataFrame(
    [], 
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns = ["{0:0.2f}".format(i / 100) for i in range(0, 10)])

for index in standardized_normal_table.index:
    for column in standardized_normal_table.columns:
        Z = np.round(float(index) + float(column), 2)
        standardized_normal_table.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

standardized_normal_table.rename_axis('Z', axis = 'columns', inplace = True)

standardized_normal_table
```

# 3 Sampling

## 3.1 Population and Sample
***

### Population
Conjunto de todos os elementos de interesse em um estudo. Diversos elementos podem compor uma população, por exemplo: pessoas, idades, alturas, carros etc.

Com relação ao tamanho, as populações podem ser limitadas (populações finitas) ou ilimitadas (populações infinitas).

### Finite Populations

Set of all the elements of interest in a study. Several elements can make up a population, for example: people, ages, heights, cars, etc.

With respect to size, populations can be limited (finite populations) or unlimited (infinite populations).

### Infinite populations

It is not possible to count its elements. As examples we have the amount of portions that can be extracted from sea water for an analysis, temperature measured at each point in a territory, etc.

When the elements of a population can be counted, but presenting a very large quantity, the population is assumed to be infinite.

### Sample
Representative subset of the population.

The numerical attributes of a population such as its mean, variance and standard deviation are known as **parameters**. The main focus of statistical inference is precisely to generate estimates and test hypotheses on population parameters using sample information.

## 3.2 When to use a sample?
***

### Infinite Population

The study would never come to an end. It is not possible to investigate all members of the population.

### Destructive testing

Studies where the evaluated elements are totally consumed or destroyed. Example: life tests, crash safety tests on cars.

### Quick results

Research that needs more agility in disclosure. Example: opinion polls, surveys that involve public health problems.

### High costs

When the population is finite but very large, the cost of a census can make the process unworkable.

## 3.4 Stratified Sampling
***

It is an improvement of the simple random sampling process. In this method it is proposed to divide the population into subgroups of elements with similar characteristics, that is, more homogeneous groups. With these separate subgroups, the simple random sampling technique is applied within each subgroup individually.

## 3.5 Sampling by Conglomerates
***


It also aims to improve the simple random sampling criterion. In conglomerate sampling subgroups are also created, but they will not be homogeneous as in stratified sampling. In conglomerate sampling, subgroups will be heterogeneous, where simple or stratified random sampling will then be applied.

A very common example of the application of this type of technique is in the division of the population into territorial groups, where the elements investigated will have quite varied characteristics.
