```python
import pandas as pd
```

```python
df = pd.read_csv("data/aluguel.csv", sep=";")
df
```

# Count the amount of properties that hat between 1 and 2 Quartos, 3 and 4, 5 and 6, >= 7 

```python
ranges = [0, 2, 4, 6, 100]
```

```python
quartos_ranges = pd.cut(df["Quartos"], bins=ranges)
quartos_ranges
```

```python
pd.value_counts(quartos_ranges)
```

```python
labels = ['1 e 2 Quartos', '3 e 4 Quartos', '5 e 6 Quartos', '>= 7 Quartos']
quartos_ranges = pd.cut(df["Quartos"], bins=ranges, labels=labels)
quartos_ranges
```

```python
pd.value_counts(quartos_ranges)
```
